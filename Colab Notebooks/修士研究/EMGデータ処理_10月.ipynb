```python
import pandas as pd
from scipy.signal import butter, filtfilt
```

```python
# ファイルを読み込む
file_path = '/content/drive/MyDrive/Colab Notebooks/修士研究/左腕/20241019金山君正規化前/mivib並列データ収集CSV_1019130511_演算.csv'
```

```python
df = pd.read_csv(file_path, encoding='latin-1')  # or try 'iso-8859-1' or 'cp1252'
```

```python
df.head(18)
```

```python
print(df.columns[1:4])
```

```python
# 第11行目以降のデータを抽出（0-indexなので10行目以降）
filtered_datapre = df.iloc[10:].copy()
```

```python
# 第2～4列目にフィルタを適用
filtered_data = filtered_datapre.iloc[:, 1:4]
```

```python
filtered_data.head(18)
```

```python
# サンプリング周波数とフィルタの設定
fs = 200  # サンプリング周波数 (Hz)
cutoff = 0.05  # カットオフ周波数 (Hz)
order = 2  # フィルタの次数
```

```python
# 正規化カットオフ周波数
nyq = fs / 2  # ナイキスト周波数
normal_cutoff = cutoff / nyq
```

```python
# Import the necessary module
import scipy.signal as signal
```

```python
# ハイパスバターワースフィルターの設計
b, a = signal.butter(order, normal_cutoff, btype='high', analog=False)
```

```python
# Assuming filtered_data contains the column names you want to filter
filtered_data_columns = filtered_data.columns  # Get the column names from filtered_data

# Loop through the columns you want to filter
for col in filtered_data_columns:
    # Convert the column to numeric, handling errors
    filtered_datapre[col] = pd.to_numeric(filtered_datapre[col], errors='coerce')

# Fill NaN values if necessary
filtered_datapre[filtered_data_columns] = filtered_datapre[filtered_data_columns].fillna(0)  # Replace 0 with an appropriate fill value

# Apply the filter to the selected columns
filtered_datapre[filtered_data_columns] = signal.filtfilt(b, a, filtered_datapre[filtered_data_columns], axis=0)
```

```python
# 正規化とフィルタ適用後のデータを確認
filtered_datapre.head()
```

```python
# CSVファイルとして保存
filtered_datapre.to_csv('1019mivibafter.csv', index=False)
```
