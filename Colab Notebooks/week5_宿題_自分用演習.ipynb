例: `X_column = 'volatile acidity'`、`Y_column = 'quality'`のとき、決定係数は$0.1525\cdots$です。

```
# common
import numpy as np
import pandas as pd
from sklearn import linear_model

# URL
path_winequality_data = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
```

```
# working place. everything 
# working place. everything
def homework(path_winequality_data, X_colmun, Y_colmun):

    wine=pd.read_csv(path_winequality_data,sep=';')# read.csvでデータを読み込む

    reg = linear_model.LinearRegression()# 線形回帰のインスタンスを生成

    X = wine[[X_colmun]].values# 説明変数X_colmun

    Y = wine[Y_colmun].values# 目的変数Y_colmun

    reg.fit(X, Y)# 予測モデルを計算

    my_result = reg.score(X,Y)# 決定係数を計算
    return my_result
```
