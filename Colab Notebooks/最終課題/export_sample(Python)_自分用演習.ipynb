# 最終課題・csvデータ取得サンプルコード

最終課題のデータを取得するための参考資料です。  
以下を実行することで、2つのテーブルより、全てのデータを取得することができます。  
**※ Pythonの標準ライブラリを用います。**

```python
import sqlite3
import csv
```

ドライブをマウントし、データベースファイル「telecom.sqlite3」のパスを指定します。

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dbpath = '/content/drive/MyDrive/Colab Notebooks/最終課題/database/telecom.sqlite3'   # データベースのパスを指定するように適宜変更する
```

データベースに接続します。

```python
conn = sqlite3.connect(dbpath)
```

以降、テーブルに対する操作を順に実行してください。

```python
# カーソルの取得
cur = conn.cursor()
```

```python
# テーブル名の確認
cur.execute('select name from sqlite_master where type = "table"')
for row in cur.fetchall():
  print(row)
```

```python
# Clientテーブルから全データを取得
cur.execute('select * from Client')
with open('Client1.csv','w',newline = '') as csv_file:
  csv_writer = csv.writer(csv_file)
  csv_writer.writerow([i[0] for i in cur.description])
  csv_writer.writerows(cur)
```

```python
# Recordテーブルから全データを取得
cur.execute('select * from Record')
with open('Record1.csv','w',newline = '') as csv_file:
  csv_writer = csv.writer(csv_file)
  csv_writer.writerow([i[0] for i in cur.description])
  csv_writer.writerows(cur)
```

```python
# 処理の確定
conn.commit()
```

```python
# カーソルを閉じる
cur.close()
```

```python
# 接続を閉じる
conn.close()
```
